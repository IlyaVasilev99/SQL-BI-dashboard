```python
def get_top3_by_each_year(dataframe):
    
    unique_years = np.sort(droppedna_years['Year'].unique())
    unique_years_today=unique_years[unique_years<=today_year][::-1]
    latest_year = unique_years_today[0]
    
    df_first_year = droppedna_years.where(droppedna_years['Year']==latest_year).sort_values(by=['User Rating'], ascending=False).head(3).reset_index(drop=True)
    unique_years_before=unique_years_today[1:]
    #print(unique_years_before)
    #print(latest_year)
    a=0
    for i in unique_years_before:
        #print(unique_years_before[a])
        df_next_year = droppedna_years.where(droppedna_years['Year']==unique_years_before[a]).sort_values(by=['User Rating'], ascending=False).head(3).reset_index(drop=True)
        df_first_year = pd.concat([df_first_year, df_next_year], axis=0)
        #if a==3:
        #    break
        a+=1
    return df_first_year
```

```python
top3_by_each_year=get_top3_by_each_year(droppedna_years)
```

```python
top3_by_each_year = top3_by_each_year.dropna(axis=0,how='all')
```

```python
top3_by_each_year
```

***BEST GENRE TO WIN GOTY***

```python
top3_1990_2023=top3_by_each_year.where(top3_by_each_year['Year']>=1990).dropna(axis=0,how='all')
```

```python
top3_1990_2023[['First Genre', 'Second Genre', 'Third Genre']] = top3_1990_2023['Genre'].str.split(', ', expand=True)
```

```python
top3_1990_2023
```

```python
All_genres_rating=top3_1990_2023.groupby(['First Genre', 'Second Genre', 'Third Genre'], as_index=True).agg(User_rating=('User Rating', 'mean'))
All_genres_rating.sort_values('User_rating', ascending=False)
```

```python
first_genres_rating=top3_1990_2023.groupby(['First Genre']).agg(User_rating=('User Rating', 'mean'),
                                                                                             genre_count=('First Genre', 'count'))
first_genres_rating
```

```python
second_genres_rating=top3_1990_2023.groupby(['Second Genre']).agg(User_rating=('User Rating', 'mean'),
                                                                                             genre_count=('Second Genre', 'count'))
second_genres_rating
```

```python
third_genres_rating=top3_1990_2023.groupby(['Third Genre']).agg(User_rating=('User Rating', 'mean'),
                                                                                             genre_count=('Third Genre', 'count'))
third_genres_rating
```

```python
all_genres_1level=pd.concat([first_genres_rating, pd.concat([second_genres_rating, third_genres_rating], join="inner")], join="inner")
```

```python
all_genres_1level.insert(0, "weight_of_rating", (all_genres_1level['User_rating']*all_genres_1level['genre_count']).round(5), True)
```

```python
all_genres_1level
```

```python
best_genres_weighted=all_genres_1level.groupby(all_genres_1level.index).agg(
                                                       genre_count=('genre_count','sum'),
                                                       weight_of_rating=('weight_of_rating','sum'))
best_genres_weighted
```

```python
best_genres_weighted.insert(0, "Weighted Rating", (best_genres_weighted['weight_of_rating']/best_genres_weighted['genre_count']), True)
```

```python
best_genre_to_win_goty = best_genres_weighted.drop(['genre_count', 'weight_of_rating'], axis=1).sort_values(['Weighted Rating'], ascending=False)
```
